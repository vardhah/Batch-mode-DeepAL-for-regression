```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from operator import add
import seaborn as sns
```

```python
uni_data= np.loadtxt("./sim_result_uni_random_batch20.txt", delimiter=",",skiprows=0, dtype=np.float32)
print(uni_data.shape)
max_uni=uni_data.max()
max_uni
```

```python
l_st=400
```

```python
epg_data= np.loadtxt("./sim_result_e_greedy_batch20.txt", delimiter=",",skiprows=0, dtype=np.float32)
print(epg_data.shape)
epg_data.max()
```

```python
index=np.arange(1,(l_st+1),dtype=int)
```

```python
fig,ax=plt.subplots(figsize=(10,6))
#plt.plot(quotients,c='b',label='ST',linewidth=3.0)
plt.plot(index,uni_data,c='black',label='random',linewidth=3.0)
plt.plot(index,epg_data,c='red',label='ep_greedy',linewidth=3.0)
plt.legend(fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=15)
plt.title('% age Accuracy on sampling methods (batch =20,itr=400)',fontsize=16,weight='bold')
plt.hlines(max_uni,170,400,linestyles= 'dashed')
#plt.hlines(max_uni,6,25,linestyles= 'dashed')
plt.vlines(170,0,max_uni,linestyles= 'dashed')
#plt.vlines(5.5,0,max_lhc,linestyles= 'dashed')
#plt.xticks(np.arange(0, 25+1, 1.0))
#plt.xlim(0,26)
plt.ylim(0.05,1.0)
plt.grid()
plt.xlabel('Num of samples=index*20',fontsize=20,weight='bold')
plt.ylabel('Accuracy',fontsize=20,weight='bold')
plt.text(163,0.01,'170',color='red',fontsize=12,weight='bold')
plt.show()
```
